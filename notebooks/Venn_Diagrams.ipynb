# Venn Diagram generation

```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

# Define set sizes and intersections
venn = venn3(
    subsets={
        '100': 20,  # AI only
        '010': 30,  # CS only
        '001': 25,  # Math only
        '110': 15,  # AI ∩ CS
        '101': 10,  # AI ∩ Math
        '011': 12,  # CS ∩ Math
        '111': 8    # AI ∩ CS ∩ Math
    },
    set_labels=('Artificial Intelligence', 'Computer Science', 'Mathematics')
)

# Customize the appearance
for idx, subset in enumerate(venn.subset_labels):
    if subset:  # Check if the label exists
        subset.set_fontsize(12)

# Set title and layout
plt.title("Venn Diagram: AI, Computer Science, and Mathematics")
plt.show()
```

```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

# Define set sizes and intersections
venn = venn3(
    subsets={
        '100': 20,  # AI only
        '010': 30,  # CS only
        '001': 25,  # Math only
        '110': 15,  # AI ∩ CS
        '101': 10,  # AI ∩ Math
        '011': 12,  # CS ∩ Math
        '111': 8    # AI ∩ CS ∩ Math
    },
    # set_labels=('Artificial Intelligence', 'Computer Science', 'Mathematics')
)

# Remove numerical labels from the diagram
for subset in venn.subset_labels:
    if subset:  # Check if the label exists
        subset.set_visible(False)

# Move set labels to the center of each circle
venn.get_label_by_id('100').set_text('AI')
venn.get_label_by_id('010').set_text('CS')
venn.get_label_by_id('001').set_text('Math')

# Customize font size and alignment for labels in the center
for id_ in ['100', '010', '001']:
    label = venn.get_label_by_id(id_)
    if label:
        print(f"Label {id_}: {label.get_text()}")
        label.set_fontsize(12)
        label.set_horizontalalignment('center')
        label.set_verticalalignment('center')

# Set title and display the diagram
plt.title("Venn Diagram: AI, Computer Science, and Mathematics")
plt.show()
```

```python
import plotly.graph_objects as go

# Define the circle properties
sets = [
    {"x": -1, "y": 0, "text": "Artificial Intelligence", "color": "rgba(255, 0, 0, 0.3)"},
    {"x": 1, "y": 0, "text": "Computer Science", "color": "rgba(0, 255, 0, 0.3)"},
    {"x": 0, "y": 1.5, "text": "Mathematics", "color": "rgba(0, 0, 255, 0.3)"},
]

# Create the figure
fig = go.Figure()

# Add circles for each set
for s in sets:
    fig.add_shape(
        type="circle",
        xref="x",
        yref="y",
        x0=s["x"] - 1,
        x1=s["x"] + 1,
        y0=s["y"] - 1,
        y1=s["y"] + 1,
        fillcolor=s["color"],
        line_color="rgba(0, 0, 0, 0)",  # Transparent border
    )
    fig.add_annotation(
        x=s["x"], y=s["y"], text=s["text"],
        showarrow=False, font=dict(size=12), align="center"
    )

# Set layout
fig.update_layout(
    title_text="Venn Diagram: AI, Computer Science, and Mathematics",
    xaxis=dict(visible=False),
    yaxis=dict(visible=False),
    showlegend=False,
    width=600,
    height=600,
)

# Display the figure
fig.show()
```
